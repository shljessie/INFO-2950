```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
```

```python
trip_df = pd.read_csv('../data/clean_trip_data.csv')
```

```python
race_income_df = pd.read_csv('../data/race_and_income_final.csv')
```

```python
start_station_geoid_trip_count = pd.DataFrame(trip_df.groupby("start_station_geoid").size())
start_station_geoid_trip_count.rename(columns={0:'num_trips_from_geoid'},inplace=True)
```

```python
end_station_geoid_trip_count = pd.DataFrame(trip_df.groupby("end_station_geoid").size())
end_station_geoid_trip_count.rename(columns={0:'num_trips_to_geoid'},inplace=True)
```

```python
race_income_trips = race_income_df.merge(start_station_geoid_trip_count, left_on='geoid',right_on='start_station_geoid')
race_income_trips = race_income_trips.merge(end_station_geoid_trip_count, left_on='geoid',right_on='end_station_geoid')
race_income_trips['num_trips_to_from_geoid'] = race_income_trips['num_trips_from_geoid'] + race_income_trips['num_trips_to_geoid']
```

```python
race_income_trips['white_household_proportion'] = race_income_trips['2017 race: White'] / race_income_trips['2017 race: total households']
race_income_trips['black_household_proportion'] = race_income_trips['2017 race: Black'] / race_income_trips['2017 race: total households']
race_income_trips['asian_household_proportion'] = race_income_trips['2017 race: Asian'] / race_income_trips['2017 race: total households']
```

```python
race_income_trips.head()
```

```python
sns.relplot(x='black_household_proportion',y='num_trips_to_from_geoid',data=race_income_trips)
plt.title('Proportion of black house holds vs number of trips taken to and from the area')
plt.show()
```

```python
sns.relplot(x='white_household_proportion',y='num_trips_to_from_geoid',data=race_income_trips)
plt.title('Proportion of white house holds vs number of trips taken to and from the area')
plt.show()
```

```python
sns.relplot(x='asian_household_proportion',y='num_trips_to_from_geoid',data=race_income_trips)
plt.title('Proportion of asian house holds vs number of trips taken to and from the area')
plt.show()
```

```python
sns.relplot(x='2017 median income',y='num_trips_to_from_geoid',data=race_income_trips)
plt.title('Income of area vs number of trips taken to and from the area')
plt.show()
```

```python
predictor_vars = ['2017 median income','white_household_proportion','black_household_proportion','asian_household_proportion']
lr_model = LinearRegression()
lr_model.fit(race_income_trips[predictor_vars],race_income_trips['num_trips_to_from_geoid'])
```

```python
print("score:",lr_model.score(race_income_trips[predictor_vars],race_income_trips['num_trips_to_from_geoid']))
for i,var in enumerate(predictor_vars):
    print(var+":"+str(lr_model.coef_[i]))
```
