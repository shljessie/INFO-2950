```python
import pandas as pd 
```

```python
trips_df=pd.read_csv('../data/citibike_trip_final.csv')
```

```python
stations_df = pd.read_csv('../data/citibike_stations_final.csv')
```

```python
stations_df.head()
```

```python
stations_df = stations_df.merge(pd.DataFrame(trip_df.groupby('start_station_geoid').size()),left_on='geoid',right_on='start_station_geoid')
```

```python
stations_df.rename(columns={0:'trips_starting_here_2017'},inplace=True)
```

```python
stations_df.head()
```

```python
stations_df = stations_df.merge(pd.DataFrame(trip_df.groupby('end_station_geoid').size()),left_on='geoid',right_on='end_station_geoid')
```

```python
stations_df.rename(columns={0:'trips_ending_here_2017'},inplace=True)
```

```python
stations_df.head()
```

```python
stations_df['total_trips_2017']=stations_df['trips_ending_here_2017']+stations_df['trips_starting_here_2017']
```

```python
stations_df.drop(columns='rides',inplace=True)
```

```python
stations_df.head()
```

```python
stations_df.drop(columns='Unnamed: 0',inplace=True)
```

```python
stations_df.to_csv('../data/citibike_stations_final.csv',index=False)
```
