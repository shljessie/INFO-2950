### Filtering out Bike Accidents 

We will filter based on

- column NUMBER OF CYCLIST INJURED/ Killed that is greater than 1 

Since the we have 1121576 in the original dataset, we will reduce this to single out accidents that involve bikes.

** Though we acknowledge that sometimes there are bike related collisions where bike riders were not injured/killed at all,
(so filtering the data by  NUMBER OF CYCLIST INJURED/ KILLED >1 will fail to encompass all of the bike related accidents). But we decided on this way of filtering to focus on accidents where the safety of cyclists where directly impacted **

#### Filtering rows where NUMBER OF CYCLIST INJURED/ KILLED is greater than 1

```python
# filter columns where the number of cyclists injured & killed is greater than zero
filter_cyclist = raw_data[(raw_data['NUMBER OF CYCLIST INJURED']>0) | (raw_data['NUMBER OF CYCLIST KILLED']>0)] 
filter_cyclist.head(-1)
```

```python
filter_cyclist.shape
```

After the filtering process for cyclist related collsions we have 23734 rows and 29 columns left.

## Filtering rows with missing latitude longitude

In the next step of our data analysis we will focus on using our longitude latitude to label each collision with a geoid refering to a specific location census tract. 

Therefore it is important that all the rows in the data contain a longitude, latitude  value. Our next step is to filter our rows that are missing longitude latitude values

#### Filtering rows where Longitude or Latitude is missing.

```python
# filter out rows missing longitude latitude values
filter_long_lat =  filter_cyclist[filter_cyclist.LONGITUDE.notnull() | filter_cyclist.LATITUDE.notnull()]
filter_long_lat
```

After the filtering process for missing values in longitude latitude we have 21846 rows and 29 columns left.

## Filtering rows with missing zip code and borough

Though at this point we don't know if we will be using the zipcode and borough columns we have decide to filter out the missing values there too just incase we need to use them later on

```python
# filter out rows missing zip code and borough 
filter_zip =  filter_long_lat[filter_long_lat['BOROUGH'].notnull() | filter_long_lat['ZIP CODE'].notnull()]
filter_zip
```

```python
filter_zip.shape
```

After the filtering process for missing values in zipcode and borough we have 17153 rows and 29 columns left.
